Agrupar por ao de estreno y calificacin para obtener la duracin promedio

```python
df['duration']
```

```python
#eliminamos nulos de la columna duration
df.dropna(subset=['duration'], inplace=True)

#filtramos a peliculas
df = df[df['type'] == 'Movie']

#eliminamos min de la columna duration
df['duration'] = df['duration'].str.replace(' min', '')

#convertimos la columna duration a entero
df['duration'] = df['duration'].astype(int)

#agrupamos por ao de estreno y calificacion y le aadimos la duracion promedio
df.groupby(['release_year', 'rating'])['duration'].mean().reset_index(name='duracion promedio').round(2)
```

```python
#agrupar por pais y sumar la duracion de las peliculas
df.groupby('country')['duration'].sum().reset_index(name='duracion total')
```
